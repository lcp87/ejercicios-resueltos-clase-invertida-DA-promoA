# **¿Qué es Seaborn?** <a class="anchor" id="section_1_1"></a>

Seaborn es una librería utilizada para realizar gráficos estadísiticos en Python. Esta construida sobre las bases de **Matplotlib** y se integra de forma impecable con **pandas**.

Lo cual nos abre una serie de puertas para facilitar la creación de gráficas complejas en muy pocas líneas de código. Así mismo tiene gráficas creadas exclusivamente para la visualización de variables categóricas.

Descubriremos que podemos realizar las gráficas realizadas en las lecciones anteriores en **Matplotlib** con **Seaborn** y explicaremos nuevos gráficos

## **Gráfica lineal**  <a class="anchor" id="section_1_3_1_1"></a>

Si deseásemos realizar una gráfica lineal conectando los diferentes puntos representados para las varaibles x e y, haciendo uso de seaborn podríamos utilizar el siguiente código como sintaxis general:
```

fig = sns.lineplot(x='variable_x' , y='variable_y' , data = dataset, ci=None, hue ='opcional')
```

De esta forma en los argumentos del linepot, se deberia de especificar:

- x : Nombre de la  variable (en el dataset) a dibujar en el eje X
- y : Nombre de la  variable (en el dataset) a dibujar en el eje X
- data: Nombre del dataset
- ci: Para mostrar el intervalo de confiaza de los datos, por defecto esta activado
- hue : Por si queremos utilizar una variable semántica para diferentes categorías


````python
plt.figure(figsize =(8,4))
fig = sns.lineplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin, ci=None)
fig.set(xlabel='Masa corporal en g', ylabel = 'Longitud de la aleta en mm')
fig.set_title('Gráfica lineal')
plt.show()
````

Aquí se puede ver la relación entre los pesos en gramos y la longuitud de las aletas de todos los datos de pingüinos disponibles.

Si ahora por ejemplo quisieramos mostrar las misma gráfica, pero diferenciando por especie o por los sexos, con añadir una pequeña opción  lo anterior conseguiremos ese resultado!

Ya vereís como es mucho más rápido que utilizando Matplotlib directamente .

Para ello añadimos la instrucción

hue ='variable_para_las_categorias'

Que se encargará de utilizar dicha variable semántica para determinar los colores de los diferentes elementos representados.

````python
# Seleccionando la especie como variable semantica en el hue
plt.figure(figsize =(8,4))
fig = sns.lineplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin, ci=None, hue ='species')
fig.set(xlabel='Masa corporal en g', ylabel = 'Longitud de la aleta en mm')
fig.set_title('Gráfica lineal')
plt.show()
# Seleccionando el sexo como variable semantica en el hue
plt.figure(figsize =(8,4))
fig = sns.lineplot(x='body_mass_g' , y='bill_length_mm' , data = df_penguin, ci=None, hue ='sex')
fig.set(xlabel='Masa corporal en g', ylabel = 'Longitud de la aleta en mm')
fig.set_title('Gráfica lineal')
plt.show()
````

Como se puede ver automáticamente nos ha separado por la variable semántica utilizada y nos ha creado la leyenda de forma automática!

### **Histogramas** <a class="anchor" id="section_1_3_1_3"></a>

Si ahora queremos realizar histogramas con Seaborn podemos hacer uso de la instrucción

```python
fig = sns.histplot(x='variable_x', data = dataset,hue = 'opcional', kde=True, bins= n_bins)

```


De esta forma en los argumentos del scatterplot, se deberia de especificar:

- x : Nombre de la  variable (en el dataset) a dibujar en el eje X
- data: Nombre del dataset
- hue : Por si queremos utilizar una variable semántica para diferentes categorías
- bins: Número o lista de bins
- kde: Para mostrar la densidad de probabilidad.


````python
# Histograma seleccionando un total de 15 bins

plt.figure(figsize =(8,4))
fig = sns.histplot(x='body_mass_g', data = df_penguin, bins = 15)
fig.set(xlabel='Masa corporal en g', ylabel = 'Cuentas')
fig.set_title('Histograma de los pesos de los pingüinos' )
plt.show()


#  Histograma seleccionando como varaible semantica la variable isla y  un total de 20 bins
plt.figure(figsize =(8,4))
fig = sns.histplot(x='body_mass_g', data = df_penguin,hue = 'island', bins = 20)
fig.set(xlabel='Masa corporal en g', ylabel = 'Cuentas')
fig.set_title('Histograma de los pesos de los pingüinos, en función de la isla')
plt.show()



#  Histograma seleccionando como varaible semantica la variable isla, un total de 20 bins y con KDE 
plt.figure(figsize =(8,4))
fig = sns.histplot(x='body_mass_g', data = df_penguin,hue = 'island', kde=True)
fig.set(xlabel='Masa corporal en g', ylabel = 'Cuentas')
fig.set_title('Histograma de los pesos de los pingüinos, en función de la isla, con su densidad de probabilidad')
plt.show()
````

Como se puede ver de los ejemplos anteriores, podemos personalizar los aspectos más relevantes de la gráfica a mostrar con las diferentes argumentos a la hora de realizar los gráficos.

Esto es importante conocerlo, ya que en más de una ocasión, considerar el número de bins a representar o mostrar por la información utilizando las diferentes categorías disponibles, nos permitirá tener un mayor conocimiento de los datos representados.

## **Diagramas de caja** <a class="anchor" id="section_1_3_1_4"></a>


Si ahora queremos realizar diagramas de caja utilizando **Seaborn** podemos hacer uso de la instrucción:

```python
fig = sns.boxplot(x='variable_x', data = dataset,hue = 'opcional', kde=True, bins= n_bins)

```


De esta forma en los argumentos del scatterplot, se deberia de especificar:

- x : Nombre de la  variable (en el dataset) a dibujar en el eje X
- x : Nombre de la  variable (en el dataset) a dibujar en el eje Y
- data: Nombre del dataset
- hue : Por si queremos utilizar una variable semántica para diferentes categorías
- bins: Número o lista de bins
- kde: Para mostrar la densidad de probabilidad.

En el caso de los boxplots en Seaborn, por defecto vienen con orientacion horizontal. De forma que para que el boxplot sea representado de forma vertical, deberemos escribir el nombre de la variable a representar en el eje y de los argumentos de boxplot.

El hecho de que esto funcione así, nos permitirá facilmente representar boxplot haciendo uso de variables categóricas para ver con más detalle como están distribuidos los datos.

Lo ilustraremos con los siguentes ejemplos:

````python
# Boxplot por defecto 

plt.figure(figsize =(8,4))
fig = sns.boxplot( x ='bill_length_mm', data = df_penguin)
fig.set(xlabel='Longitud de la aleta en mm')
fig.set_title('Boxplot de la longitud de la aleta')
plt.show()
````

````python
# Boxplot por vertical 

plt.figure(figsize =(8,4))
fig = sns.boxplot( y ='bill_length_mm', data = df_penguin)
fig.set(xlabel='Longitud de la aleta en mm')
fig.set_title('Boxplot de la longitud de la aleta')
plt.show()
````

Si ahora por ejemplo deseásemos utilizar algun argumento adicional para utilizar las variables categóricas de las que disponemos en dihco dataset, podríamos utilizar estos valores para el argumento de X y del hue

````python
# Boxplot por vertical, por isla
plt.figure(figsize =(8,4))
fig = sns.boxplot(x= 'island', y ='bill_length_mm', data = df_penguin)
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm')
fig.set_title('Boxplot de la longitud de la aleta en mm por isla')
plt.show()


# Boxplot por vertical, por isla
plt.figure(figsize =(8,4))
fig = sns.boxplot(x= 'island', y ='body_mass_g',hue= 'species', data = df_penguin)
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm por isla y sexo' )
fig.set_title('Boxplot de la longitud de la aleta')
plt.show()
````

Como se puede ver, al utilizar las diferentes variables categóricas disponibles, podemos extraer información adicional a la hora de investigar los datos.
 Esto será de especial utilidad, cuando en futuras lecciones hablemos del analisis exploratorio de datos.

### **Diagrama de enjambre** <a class="anchor" id="section_1_3_2_1"></a>

Este típo de gráficas se utilizan para representar los diagrmas que enjambre, que no son más que una gráfica de dispersion para varibles categóricas, de forma que los puntos no se superpongan unos sobre otros, lo cual nos ayuda también a ver como están distribuidos los puntos en el rango de valores dichos datos.

Para hacer este tipo de gráficas, utilizaremos la siguente sintaxis:

```python
fig = sns.swarmplot(x='variable_x', y= 'variable_y', data = dataset)

```


De esta forma en los argumentos del scatterplot, se deberia de especificar:

- x : Nombre de la  variable (en el dataset) a dibujar en el eje X
- y : Nombre de la  variable (en el dataset) a dibujar en el eje Y
- data: Nombre del dataset
- hue : Por si queremos utilizar una variable semántica para diferentes categorías


````python
# Diagrama de enjambre del  peso de los pingüinos por isla
plt.figure(figsize =(8,4))
fig = sns.swarmplot(x='island', y = 'body_mass_g' ,data = df_penguin)
fig.set(xlabel='Isla', ylabel= 'Peso de los pingüinos por isla' )
fig.set_title('Diagrama de enjambre del  peso de los pingüinos por isla')
plt.show()

````

De la gráfica anterior se puede extraer a simplevista que los pesos de los pingüinos de la isla Biscoe, son por lo general más altos. Esto podría implicar que en los alrededores hay mayor cantidad de comida disponible.

Si ahora vamos un paso más alla y diferenciamos por el sexo de los pingüinos

````python
# Diagrama de enjambre del  peso de los pingüinos por isla y sexo
plt.figure(figsize =(8,4))
fig = sns.swarmplot(x='island', y = 'body_mass_g' ,hue='sex', data = df_penguin)
fig.set(xlabel='Isla', ylabel= 'Peso de los pingüinos por isla' )
fig.set_title('Diagrama de enjambre del  peso de los pingüinos por isla y sexo')
plt.show()
````

En las gráficas anteriores, haciendo uso de las varibles categóricas disponibles hemos podido entender mejor como están distribuidos los datos de los pesos de los pinguinos pingüinos, para las diferentes islas, así como los diferentes sexos. De la gráfica anterior, podemos extraer que los pingüinos macho suelen tener un peso mayor que en el caso de las hembras y que el peso medio ( donde mayor cantidad de puntos hay en las gráficas), es superior en todos los casos para ambos sexos en el caso de la isla de Biscoe.

### **Countplot** <a class="anchor" id="section_1_3_2_2"></a>

Como se ha visto hasta ahora, cuando buscamos realizar histogramas o diagramas de barras, siempre hemos tenido que trabajar con variables de tipo numérico.
No obstate en Seaborn existe una función que nos permitirá hacer esto mismo utilizando las viables categóricas del dataset con el que estemos trabajando.

Para ello podremos hacer uso de la instrucción countplot, que tiene la siguente sintaxis:

```python
fig = sns.countploy(x='island', data = dataset)

```


De esta forma en los argumentos del countplot, se deberia de especificar:
Tendremos que escoger unicamente si queremos representar los datos en el eje de las X o Y.

- x : Nombre de la  variable (en el dataset) a dibujar en el eje X
- y : Nombre de la  variable (en el dataset) a dibujar en el eje Y
- data: Nombre del dataset
- hue : Por si queremos utilizar una variable semántica para diferentes categorías.

De esta forma vamos a ilustrarlo con un ejemplo:



````python
plt.figure(figsize =(8,4))
fig = sns.countplot(x='island', data = df_penguin)
fig.set(xlabel='Isla', ylabel= 'Cuentas' )
fig.set_title('Histograma de varible categórica')
plt.show()
````

De esta forma hemos obtenido rápidamente el valor total de de pingüinos por cada una de las islas.Como si de una variable numérica se tratase.
Si ahora adicionalmente añadiéramos algún valor al atributo hue en la gráfica anterior:

````python
plt.figure(figsize =(8,4))
fig = sns.countplot(x='island', data = df_penguin, hue = 'sex')
fig.set(xlabel='Isla', ylabel= 'Cuentas' )
fig.set_title('Histograma de varible categórica')
plt.show()
````

Ahora ya conocemos con todo lujo, la cantidad de pingüinos que hay por cada sexo en cada una de las islas. Como se puede ver, tener la capacidad de representar variables categóricas es extremadamente útiles para conocer en detalle los datos que se disponen.

### **Catplot** <a class="anchor" id="section_1_3_2_3"></a>


Tendremos que escoger unicamente si queremos representar los datos en el eje de las X o Y.

- x : Nombre de la  variable (en el dataset) a dibujar en el eje X

- data: Nombre del dataset

fig.suptitle(), sobre el propio gráfico.

````python
plt.figure(figsize =(8,4))
fig = sns.catplot(x='island', y ='bill_length_mm', data = df_penguin)
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm' )
#fig.set_title('Histograma de varible categórica')
plt.show()
````

De la gráfica anterior podemos extraer la diferentes longitudes de las aletas por cada una de las islas y ver donde estarían representados cada uno de los valores en la gráfica.

Si ahora adicionalemente quisieramos saver como esta separado por especies, añadiendo el hue podríamos conseguirlo rápidamente.



````python
plt.figure(figsize =(8,4))
fig = sns.catplot(x='island', y ='bill_length_mm', data = df_penguin, hue = 'species')
fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm' )
fig.fig.suptitle('Histograma de varible categórica')
plt.show()
````

### **Mapas de calor (Heatmap)** <a class="anchor" id="section_1_3_2_5"></a>

Finalmente para terminar esta lección vamos a explicar los mapas de calor. 

Un mapa de calor nos permite mostar en una gñrfica la magnitud de una variable haciendo uso de colores. Donde la intensidad del color nos indicará la mayor o menor presencia de la magnitud representada.

Posiblemente os hayais encontrado con este tipo de gráficas en algún punto de vuestra vida, si no, esta es vuestra primera introducción. Ya vereis lo útiles que pueden llegar a ser.

La instruccón a utilizar para realizar este tipo de gráficas es la siguente:



```python
fig = sns.heatmap(data = dataset)

```

Nota: Hemos de tener en cuanta que unicamente podemos obtener el mapa de calor de aquellas varibles que sean de tipo numérico y como consideración general, se recomienda si se representa más de una variable que estas estén el mismo orden de escala, es decir de rango similares. Ya que si se representa magnitudes muy dispares, será complicado dilucidar la información mostrada.

Por ejemplo si aplicásemos el mapa de calor para el dataset con el que estamos trabajando, sin considerar lo anterior:


````python
plt.figure(figsize =(8,4))
fig = sns.heatmap(df_penguin[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']])
#fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm' )
#fig.title('Histograma de varible categórica')
plt.show()
````

Como se puede observar nos muestra como de frecuente son los valores representados en función de la intensidad del color mostrada.

No obstante, al no haber considerado las diferentes ordenes de los datos representados, no se puede comprender bien la información mostrada. Hemos juntado cantidades de decenas(información de las aletas) con millares(peso de los pingüinos)

Si ahora tomásemos varibles con rangos similares o únicamente una de las variables:


````python
plt.figure(figsize =(8,4))
fig = sns.heatmap(df_penguin[['bill_length_mm']])
#fig.set(xlabel='Isla', ylabel= 'Longitud de la aleta en mm' )
#fig.title('Histograma de varible categórica')
plt.show()
````

El potencial de esta gráfica lo descubriremos más adelante, cuando estemos explortando las difererentes relaciones que tiene unas variables con otras.
