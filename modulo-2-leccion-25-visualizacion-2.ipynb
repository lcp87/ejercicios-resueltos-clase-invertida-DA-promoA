# **Gráficas con estilos** <a class="anchor" id="section_1_3"></a>

De ahora en adelante vamos a utilizar una sintaxis general para definir las gráficas y todas las propiedades de personalización de los ejes.

De ahora en adelante, cada vez que deseemos crear una nueva gráfica haremos uso de la siguente instrucción:

```python

fig, ax = plt.subplots()

```

Que a efectos prácticos nos creará un objeto de tipo figure sobre el objeto **fig** y adicionamente podremos modificar aspectos sobre la personalizacion de los ejes, titulos, nombre de los ejes, utilizando el objeto **ax**.

Ahora no es importante conocer en detalle el funcionamiento de **plt.subplot()**, ya que lo explicaremos en detalle a lo largo de la lección.

## **Tamaños de las gráficas** <a class="anchor" id="section_1_3_1"></a>

Si ahora hacemos uso de la nueva forma de definir los objetos figura y ejes, añadiendo una sencilla instrucción adicional, podemos modificar sencillamente el tamaño de la gráfica mostrada:

Para ello tenemos que añadir lo siguente dentro de los argumentos de plt.subplots():

**figsize = (ancho, alto)**

Donde ancho y alto, seran los números en pulgadas que definan el tamaño de la gráfica.
De esta forma nos quedaría lo siguente:

```python

fig, ax = plt.subplots(figsize =(8,4))

```

Vamos a mostrar ahora como afecta los números para el tamaño y la anchura, para una gráfica ejemplo:

````python

fig, ax = plt.subplots(figsize =(8,4))

x = [10, 20, 30, 40, 50]
y = [50,100,150,200, 250]

plt.plot(x, y)
plt.show()
````

````python

fig, ax = plt.subplots(figsize =(1,4))

x = [10, 20, 30, 40, 50]
y = [50,100,150,200, 250]

plt.plot(x, y)
plt.show()
````

### **Título** <a class="anchor" id="section_1_3_3"></a>

Por lo general suele ser imporante poner un título sobre la gráfica que describa lo que se esta representando, ya que toda persona que vea la gráfica debería de saber sin mucho esfuerzo lo que se le está mostrando.

De esta forma para añadir un título a un gráfico deberemos hacer uso de la siguente instrución:

plt.title(label='titulo de la gráfica')

Vamos a ilustrar su uso con ejemplo:

````python
x =np.random.randn(100)
y = np.random.randn(100) 
fig,ax = plt.subplots(figsize= (8,4))

plt.scatter(x,y)
plt.title('Gráfica de dispersion de dos conjuntos de datos')

plt.show()
````

### **Etiquetas de los ejes** <a class="anchor" id="section_1_3_4"></a>

### **Límites de los ejes** <a class="anchor" id="section_1_3_5"></a>

Así mismo a la hora de realizar las gráficas podemos querer unicamente mostrar una parte de la gráfica, para ello vamos a limitar la longitud que mostraremos de los ejes.

Para poder realizar esto haremos uso de las siguentes instrucciones:

```python

plt.xlim([,])
plt.ylim([,])

```
Donde en [,] escribiremos el rango que deseamos mostrar.
Haciendo uso del ejemplo anterior:


````python
x =np.random.randn(100)
y = np.random.randn(100) 
fig,ax = plt.subplots(figsize= (8,4))

plt.scatter(x,y)
plt.title('Gráfica de dispersion de dos conjuntos de datos')
plt.xlim([-1,1]) # representamos unicamente el rango entre -1 y 1 en el eje x
plt.ylim([0,3]) # representamos unicamente el rango entre 0 y 3 en el eje y
ax.set_xlabel('Nombre del eje X')
ax.set_ylabel('Nombre del eje Y')
plt.show()
````

Como se puede observar ahora, al modificar los límites de la gráfica hemos dejado de poder observar los datos fuera de los rangos seleccionados para ambos ejes. Es como si hubieramos hecho zoom en la imagen.

### **Cuadrícula** <a class="anchor" id="section_1_3_6"></a>

Si ahora deseamos mostar en el fondo de la imagen una cuadrícula, para hacer más sencilla la interpretacion de las escalas de los ejes, podemos añadir la siguente instrucción a la hora de realizar la gráfica:

```python

ax.grid()

```

Donde podemos incluir algunas preferencias tales como:
- Color
- Tipo de línea
- Sobre que ejes queremos mostar la cuadrícula
- Tamaño
- etc

Para ello añadiriamos estos argumentos sobre la instrucción utilizada a la hora de representar la cuadrícula:

```python

ax.grid(color = '' , linestyle= '',linewidth = 2)

```
Donde:
- color = ''  Sería para definir el color que deseamos, introducir la primera letra del color en ingles o en nombre completo del color
    Ej: 
    - Rojo : 'r' 
    - Azul : 'b'
    - Verde : 'g'
    - Violeta : 'violet'

- linestyle= ''  Para mostar el tipo de línea que deseamos dibujar. Algunas de las opciones disponibles serían:

    Ej:
    - \'-' or 'solid' Linea solida

    - \'--' or 'dashed' Línea punteada

    - \'-.' or 'dashdot' Linea solida y con puntos


    - \':' or 'dotted' Línea punteada





- linewidth = 2 Valor numérico para la anchura de la línea.

Para ver en detalle todas las opciones de configuración de los ejes:

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html?highlight=grid



````python
x =np.random.randn(100)
y = np.random.randn(100) 
fig,ax = plt.subplots(figsize= (8,4))

plt.scatter(x,y)
plt.title('Gráfica de dispersion de dos conjuntos de datos')
plt.xlim([-1,1])
plt.ylim([0,3])

ax.set_xlabel('Nombre del eje X')
ax.set_ylabel('Nombre del eje Y')
ax.grid()
plt.show()
````

````python
# Si deseásemos incluir algunas de las opciones de personalizacion de la cuadrícula:

x =np.random.randn(100)
y = np.random.randn(100) 
fig,ax = plt.subplots(figsize= (8,4))

plt.scatter(x,y)
plt.title('Gráfica de dispersion de dos conjuntos de datos')
plt.xlim([-1,1])
plt.ylim([0,3])

ax.set_xlabel('Nombre del eje X')
ax.set_ylabel('Nombre del eje Y')
ax.grid(color ='violet', linestyle = '-', linewidth = 1)
plt.show()
````

### **Leyenda** <a class="anchor" id="section_1_3_8"></a>

La leyenda de una gráfica nos permite identificar con claridad los diferentes conjuntos de datos representados en una gráfica. Esto resulta extremadamente útil cuando estamos representando más de un conjunto de datos sobre una misma gráfica, ya que de un vistazo nos permite identificar los diferentes grupos, varibles o elementos representandos.

La sintaxis general para incluir una leyenda en un gráfico es la siguente:


```python
ax.legend()
```

De forma automática debería listar los nombres de las varibles representadas, no obstante si se desea tener mayor control, se puede escribir el nombre de las etiquetas que se desean representar añadiendo lo siguiente:

```python
ax.legend(labels =['label1','label2',...])
```

Nota: En la anterior lección ya se utilizó este argumento, sin hacer hinacapié en el, para cuando se mostraron los diagramas barras apiladas y cuando se mostraron dos diagramas de barras lado a lado.


Vamos a ilustar como añadir la leyenda haciendo uso de algunos ejemplos:



````python
x= ['Suspenso','Aprobado', 'Notable','Sobresaliente'] # edades
fig, ax = plt.subplots()
y = [1,5,4,2] # numero de personas por esa edad
y2 = [7,2,3,6]


plt.bar(x,y, label = 'grupo A')
plt.bar(x,y2, bottom= y,label = 'grupo B')
ax.legend()
plt.show()
````

En este caso la leyenda se ha completado automaticamente tomando como etiquedas, la propia etiqueda incluida en la gráfica de barras.

Vamos ahora a mostrar otro ejemplo, donde definiremos de forma manual las etiquetas que deseamos mostrar en la leyenda.

````python
fig,ax = plt.subplots(figsize= (8,4))

x = [10, 20, 30, 40, 50]
y = [50,100,150,200, 250]

x2 = [20, 30, 40, 50, 60]
y2 = [20,80,100,200, 250]

plt.plot(x, y, c= 'r')
plt.plot(x2, y2, c= 'blue')
plt.title('Gráfica lineal de dos conjuntos de datos')


ax.set_xlabel('Nombre del eje X')
ax.set_ylabel('Nombre del eje Y')
ax.legend(labels=['Recta 1','Recta 2'])
plt.show()
````

### **Colores** <a class="anchor" id="section_1_3_8"></a>


El atributo de color nos permite cambiar los colores de los objetos representado, por lo general para hacer uso de este atributo deberemos añadir en la función del tipo de gráfica que estemos realizando el atributo siguente:

c (or color) = 'Nombre_del_color'

Vamos a ver como especificaríamos este atributo para algunas de las gráficas que explicamos en la lección anterior:
```python
- plot(x, y, color ='r') o de forma alternativa plot(x, y, c ='r')
- bar(x, y, color ='blue')
- scatter(x,y, c(or color)='yellow')
- hist(x, color ='green')
```

Mostramos ahora el ejemplo de como colorear un gráfico de tipo plot:


````python
fig,ax = plt.subplots(figsize= (8,4))

x = [10, 20, 30, 40, 50]
y = [50,100,150,200, 250]

plt.plot(x, y, c= 'r')

plt.title('Gráfica lineal de dos conjuntos de datos')


ax.set_xlabel('Nombre del eje X')
ax.set_ylabel('Nombre del eje Y')
plt.show()
````

## Marcadores <a class="anchor" id="section_1_3_9"></a>

Finalmente vamos a hablar ahora de los marcadores, siendo estos el icono que representa los datos dibujados, por ejemplo en los diagramas de dispersión por defecto el marcador es un circulo, como se ha podido ver en los ejemplos anteriores.

Los tipos de marcadores son personalizables para las gráficas de tipo:
- plot
- scatter
- errorbar (No se ha explicado, se utiliza para mostar la barras de error de los datos)

Únicamente tiene sentido para aquellas gráficas en la que representemos los diferentes puntos de los datos. Por ejeplo, seria incoherente querer mostrar los marcadores en una gráfica de barras o un diagrama de sectores.

Por un  lado podemos personalizar el tamaño del marcador, haciendo uso de la siguiente instruccion, dentro del tipo de gráfica que estemos realizando:

- plot(x, y, markersize= valor_numérico ) 
- scatter(x,y,markersize= valor_numérico)


Por otro, si queremos personalizar el tipo de marcador, tomando un marcador de la siguente lista:


|Marcador| Descripción|
|----|-----|
|"."|Punto|

## **Multigráficas** <a class="anchor" id="section_1_4"></a>

Hasta ahora hemos explicado como realizar gráficas de forma independientes una de otras, pero Matplotlib es muy versatil y también nos permite incluir en la misma figura más de una gráfica.

Para ello vamos a explicar ahora en más detalle como funciona la instrucción de 
```python
plt.subplot()
```
Que hemos venido utilizando hasta ahora, sin entender exactamente su funcionamiento.

De esta forma en subplot podemos incluir lo siguente para mostrar más de una gráfica:

```python
plt.subplot((numero_filas, numero_columna,indice), fizsize= (ancho,alto)))
```

Por ejemplo si quisieramops representar 2 gráficas una al lado de la otra, podríamos escribir lo siguente:

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplot((2, 2), fizsize= (8,4)))
```

Así mismo podríamos querer personalizar las etiquetas, títulos, etc de cada una de las gráficas.

Vamos a ilustrar esto con un ejemplo:




````python
fig,ax = plt.subplots(1,2)

x = [10, 20, 30, 40, 50]
y = [50,100,150,200, 250]

x2 = [20, 30, 40, 50, 60]
y2 = [20,80,100,200, 250]


ax[0].plot(x, y, c= 'r', marker= 'X' )
ax[0].set_title('Datos 1')
ax[0].set_xlabel('Nombre del eje X')
ax[0].set_ylabel('Nombre del eje Y')
ax[0].legend(labels=['Recta 1','Recta 2'])


ax[1].plot(x2, y2, c= 'blue', marker ='D')
ax[1].set_xlabel('Nombre del eje X')
ax[1].set_ylabel('Nombre del eje Y')
ax[1].legend(labels=['Recta 1','Recta 2'])
ax[1].set_title('Datos 2')
````

## **Exportar figuras** <a class="anchor" id="section_1_5"></a>

Para finalizar la lección vamos a explicar como exportar las imágenes creadas, para poder introducirlas en un informe, mostrarlas en un powerpoint o lo que se desee hacer con ella.

Para ello tenemos un par de formas de realizarlo:

- Método 1: Guardado manual. Al poner el cursor sobre la gráfica, veremos que aparece a la izquerda del todo de la celda el icono de guardar. Haz click, escoge el nombre la gráfica y guardala.
- Método 2: Al final de la imagen añade la siguente línea:

```python
plt.savefig(nombre_de_la_figura.extension)
```
Las extensiones pueden ser:
- png
- jpg
- jpeg
- tiff
Te guardará la imagen en la misma ruta que tengas guardado el Jupyter notebook que estas ejecutando.

````python
fig,ax = plt.subplots(1,2)

x = [10, 20, 30, 40, 50]
y = [50,100,150,200, 250]

x2 = [20, 30, 40, 50, 60]
y2 = [20,80,100,200, 250]


ax[0].plot(x, y, c= 'r', marker= 'X' )
ax[0].set_title('Datos 1')
ax[0].set_xlabel('Nombre del eje X')
ax[0].set_ylabel('Nombre del eje Y')
ax[0].legend(labels=['Recta 1','Recta 2'])


ax[1].plot(x2, y2, c= 'blue', marker ='D')
ax[1].set_xlabel('Nombre del eje X')
ax[1].set_ylabel('Nombre del eje Y')
ax[1].legend(labels=['Recta 1','Recta 2'])
ax[1].set_title('Datos 2')

plt.savefig('tirori.tiff')
````
