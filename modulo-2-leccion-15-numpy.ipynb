# ¬øQu√© es NumPy? 

NumPy (**Num**erical **Py**thon) es una librer√≠a de Python que introduce los vectores y matrices en Python. Concretamente esta especializada en el c√°lculo num√©rico y el an√°lisis de datos. 

NumPy tiene su propia estructura de datos incorporada llamado *arrays* que es similar a las listas de Python, pero puede almacenar y operar con datos de manera mucho m√°s eficiente.

üìå Como analistas de datos pasaremos mucho m√°s tiempo trabajando con Pandas que con NumPy. A√∫n as√≠, Pandas se basa en NumPy, por lo que es esencial comprender los aspectos m√°s importantes de la librer√≠a NumPy.


 > En conclusi√≥n, es una librer√≠a bastante amplia y unida a Pandas son muy potentes y nos van a permitir realizar una gran cantidad de operaciones.


**CARACTER√çSTICAS DE NumPy**

- Est√° escrito en C, lo que le proporciona de una velocidad muy alta, y cuando trabajamos con grandes conjuntos de datos la velocidad es un punto fundamental.


- Incluye funciones para operaciones matem√°ticas, de l√≥gica, de ordenaci√≥n, estad√≠sticas, de entrada y salida para leer y escribir ficheros, etc√©tera.


- El objeto ‚Äúndarray‚Äù o ‚Äúarray‚Äù es el tipo de dato m√°s importante en NumPy. 

## **¬øQu√© es un array?**

Como hemos dicho al inicio, los *arrays* tambi√©n llamados "arreglos num√©ricos multidimensionales" son parecidos a las listas de Python pero con algunas diferencias importantes:

- Solo permite datos de un mismo tipo


- Son menos pesados


- Son m√°s r√°pidos


- Permiten crear f√°cilmente arrays n-dimensionales

Por lo tanto, un array es una estructura de datos de un mismo tipo organizada en forma de tabla de n-dimensiones. 


**¬øN-dimensiones? ¬øQu√© significa eso?** ü§î

Las dimensiones de un array tambi√©n se conocen como ejes. Podemos clasificarlos en tres_ 

- **unidimensional** o **vector**: es aquel en el cual se puede acceder a cualquier elemento solamente con un √≠ndice. Se puede imaginar como si se tratase de una lista de datos en la cual, para referirnos a cada elemento de dicha lista, emplearemos un n√∫mero (√≠ndice) para indicar la posici√≥n en la que podemos encontrar dicho elemento dentro de ella.

![image.png](https://github.com/Adalab/data_imagenes/blob/main/Modulo-2/NumPy/unidimensional.png?raw=true)

- **bidimensional**  o **matriz**: es aquel que necesitamos dos √≠ndices para localizar cualquier elemento. Cualquier elemento dentro del array bidimensional se puede imaginar como un elemento que se encuentra localizado dentro de una posici√≥n determinadas en un ‚Äú**eje X**‚Äù y un ‚Äú**eje Y**‚Äù.

![image.png](https://github.com/Adalab/data_imagenes/blob/main/Modulo-2/NumPy/bidimensional.png?raw=true)

En la imagen anterior, si cogemos cualquier posici√≥n dentro del array como una posici√≥n {x,y}:

    - {0,0} corresponde al n√∫mero 1 azul.
    
    - {2,3} corresponde al n√∫mero 1 rojo.
    
    - {4,4} corresponde al n√∫mero 2 verde.

- **multidimensional**: son *arrays* que tienen dos o m√°s dimensiones. En este caso el n√∫mero de √≠ndices viene determinado por el n√∫mero de dimensiones de nuestro array, por ejemplo, para un array tridimensional el n√∫mero de √≠ndices ser√≠a 3.

![image.png](https://github.com/Adalab/data_imagenes/blob/main/Modulo-2/NumPy/tridimensional.png?raw=true)

Si cogemos como ejemplo la imagen anterior, se trata de un array tridimensional, en el que, cualquier elemento, puede ser localizado mediante una posici√≥n {x,y,z}.

Por ejemplo:

    - Para el cuadrado rojo que est√° en la cara inferior izquierda la posici√≥n ser√≠a {0,2,0}
    
    - Si queremos localizar el cuadrado verde de la cara inferior derecha ser√≠a {2,1,1}

## Antes de empezar

NumPy nos va a permitir crear *arrays* con n√∫meros aleatorios con unas dimensiones dadas. Para ello tendremos que usar el m√©todo `random` que encontraremos dentro de la librer√≠a NumPy. Veamos algunos ejemplos:

1Ô∏è‚É£ `np.random.randint()`: nos devuelve n√∫meros enteros aleatorios entre dos valores que le especifiquemos. La sintaxis ser√≠a:


2Ô∏è‚É£ `np.random.rand()`: nos devuelve un *array* con la forma que le especifiquemos. La sintaxis ser√≠a:

```python

# donde d ser√°n las dimensiones de nuestro array
```

üìù **NOTA**: en este caso no tenemos que especificar que entre que valores queremos que nos genere los n√∫meros. POR DEFECTO NOS GENERAR√Å N√öMEROS ALEATORIOS ENTRE EL 0-1


üìù NOTA: Solo nos dar√° floats no hace falta especificar el tipo

## `np.eye()`

Nos permite crear un *array* con unos en una diagonal

Una **matriz de indentidad** es una matriz con ceros en las filas y columnas excepto en la diagonal que tendremos unos

üö® En este caso los valores pasados especifican el n√∫mero de filas y columnas que tendremos en nuestro *array*. 

````python
# vamos a crear una matriz de 4 filas x 4 columnas
# por defecto nos genera los unos en la diagonal principal

array17 = np.eye(4,4)
array17
````

````python
# creamos una matriz de identidad de 2 filas x 2 columnas
# para cambiar la posici√≥n de la diagonal usamos el par√°metro k 
# valores negativos mueve la diagonal hacia abajo
# el valor de k indica cuantos niveles queremos subir o bajar en la diagonal

array18 = np.eye(4,3, k = -1)
array18
````

````python
# creamos una matriz de indentidad de 2 filas x 2 columnas
# valores positivos mueve la diagonal hacia arriba

array19 = np.eye(4,3, k = 1)
array19
````

## `np.identity()`


Nos permite crear una **matriz de identidad**.

Una **matriz de identidad** es una matriz con ceros en las filas y columnas excepto en la diagonal que tendremos unos


üö® En este caso el valor pasado corresponde con el n√∫mero de filas y columnas que queremos en nuestro array

````python
# vamos a crear una matriz identidad de 4 filas x 4 columnas

array20 = np.identity(4)
array20
````

üö® La diferencia entre `np.eye` y `np.identity` es que la primera nos va a permitir desplazar la diagonal mientras que la segunda solo rellena la diagonal principal.

Lo podemos hacer de dos formas:

De la misma forma que en al suma podemos hacerlo de dos formas:
