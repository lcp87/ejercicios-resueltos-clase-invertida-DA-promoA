# Funciones Universales

Las funciones universales en Numpy son funciones matem√°ticas simples. NumPy nos va proporcionar varias funciones universales que cubren una amplia variedad de operaciones.


Estas funciones incluyen funciones trigonom√©tricas est√°ndar, funciones para operaciones aritm√©ticas, manejo de n√∫meros complejos, funciones estad√≠sticas, etc. 

Las principales car√°cteristicas de estas funciones son: 

- Operan en cualquier tipo de *array* (uni, bi o multidimensional). 


- Realizan operaciones r√°pidas.


- Numpy, las funciones universales son objetos que pertenecen a la clase numpy.ufunc.


- Algunas ufuncs se llaman autom√°ticamente cuando se usa el operador aritm√©tico correspondiente en matrices. Por ejemplo, cuando la suma de dos matrices se realiza por elementos utilizando el operador '+', se llama internamente a np.add(), las cuales vimos en la primera sesi√≥n de NumPy


A lo largo de esta leccion veremos algunas de ellas: 

üìå A modo de recordatorio, en la primera sesi√≥n vimos algunas de estas funciones, como por ejemplo:

- `add()`: para sumar


- `multiply()`: para multiplicar


- `substract()`: para restar


- `divide()`: para dividir




>üö®‚ö†Ô∏è **PERO RECORDEMOS QUE SI QUERMOS HACER OPERACIONES ENTRE DOS *ARRAYS*  DEBEN TENER LA MISMA FORMA PARA PODER HACER ESTAS OPERACIONES** üö®‚ö†Ô∏è

üö® **NOTA** A lo largo de este jupyter trabajaremos con dos *arrays* bidimensionales, pero es igualmente aplicable a cualquier tipo de *array*.

# Operaciones estad√≠sticas

## `np.mean()`

Calcula la media de todos los elementos del *array*. 

Este m√©todo nos permite elegir el eje sobre el que queremos calcular la media. Para eso tendremos que usar el par√°metro `axis`, donde:

- `axis = 0`: nos calcular√° la media para cada **columna** de nuestro *array*



- `axis = 1`: nos calcular√° la media para cada **fila** de nuesrto *array*




Cuando pasamos este par√°metro la funci√≥n `np.mean` produce un nuevo *array* con los resultados. 

````python
# si no especificamos nada nos calcular√° la media de todos los elementos del array

media = np.mean(arr1)
media
````

````python
# media por columna. Nos va a devolver un valor por cada columna de nuestro array

media_filas = np.mean(arr1, axis = 0)
media_filas
````

````python
# media por fila, nos devuelve un valor por cada fila de nuestro array

media_columnas = np.mean(arr1, axis = 1)
media_columnas
````

## `np.std()`:

Calcula la desviaci√≥n est√°ndar en nuestro *array*, por defecto nos lo va a calcular para todos los elementos. 

Si le pasamos el par√°metro `axis` podemos especificar sobre que eje queremos que nos calcule la desviaci√≥n est√°ndar. 

- `axis = 0`: columnas


- `axis = 1`: filas


>üìå **NOTA**: La desviaci√≥n est√°ndar es la medida de dispersi√≥n m√°s >com√∫n, que indica como de dispersos est√°n los datos con respecto a la >media. 
>
>Cuanto mayor sea la desviaci√≥n est√°ndar, mayor ser√° la dispersi√≥n de los >datos.

````python
# desviaci√≥n est√°ndar de media

np.std(arr1)
````

````python
# ¬ø qu√© pasa con las columnas? 

np.std(arr1, axis = 0)
````

Aunque esto lo veremos en m√°s detalle m√°s adelante, intentemos interpretar estos resultados. 

Para la primera columna tenemos el siguiente valor: 0.14085392. Esto significa que los valores de la columna 1 se dev√≠an 0.14 del valor de la media. 

````python
# ¬øy con las filas? 

np.std(arr1, axis = 1)
````

## `np.var()`:

Calcula la varianza de los valores de mi *array*. Por defecto lo har√° sobre todo el *array*, pero nosotros le podemos especificar el eje usando el par√°metro `axis`. 

- `axis = 0`: columnas


- `axis = 1`: filas


>üìå **NOTA**: es una medida de dispersi√≥n que representa la variabilidad de una serie de datos respecto a su media. Formalmente se calcula como la suma de los residuos al cuadrado divididos entre el total de observaciones.

````python
np.var(arr1)
````

## `np.cumsum()`:

Devuelve un *array* con la suma acumulada de los elementos a lo largo nuestro *array*. Por defecto nos lo har√° para todo el *array*. 


````python
np.cumsum(arr1)
````

¬øQu√© esta haciendo `np.cumsum`?

- El primer valor del *array* que nos devuelve ser√° igual al primer elemento de nuestro *array* original. 


- El segundo valor del *array* es la suma del primer elemento del original m√°s el segundo

```python

## array original
array([[0.43158982, 0.42402593, 0.29239649, 0.11322166],
       [0.62065375, 0.028709  , 0.43423764, 0.60455712],
       [0.47499132, 0.90050962, 0.54005677, 0.41098125]])

#operaci√≥n
np.cumsum(arr1)

#output de cumsum
array([0.43158982, 0.85561575, 1.14801225, 1.26123391, 1.88188766,
       1.91059666, 2.3448343 , 2.94939141, 3.42438273, 4.32489235,
       4.86494912, 5.27593037])



# el segundo elemento es 0.85561575 que es el resultado de sumar los dos primeros elementos
0.43158982 + 0.42402593

```
- El tercer valor del *array* resultado de `np.cumsum` es la suma del tercer elemento y el resultado de la suma de los dos anteriores. Y as√≠ sucesivamente hasta que se sumen todos los elementos de forma acumulativa. 


- El √∫ltimo elemento del *array* que nos devuelve ser√° igual que el resultado de hacer la suma de todos los elementos del *array*

Si nos fijamos lo que esta haciendo es coger los dos primeros elementos de nuestro *array* y los suma: 



Al igual que con la suma podemos pasarle el par√°metro `axis` para especificar sobre que eje queremos aplicar la operaci√≥n:


- `axis = 0`: columnas


- `axis = 1`: filas


````python
# para las columnas

np.cumsum(arr1, axis = 0)
````

````python
# para las filas

np.cumsum(arr1, axis = 1)
````

# Operaciones de comparaci√≥n 

## `np.any()`

Comprueba si **alg√∫n** elemento del *array* se eval√∫a como True dada una condici√≥n.

Nos va a devolver un valor booleano (True, False) a no ser que especifiquemos un eje con el par√°metro `axis`. En este caso, nos devolver√° un valor True o False en funci√≥n de si se cumple o no la condici√≥n. 

Recordamos: 

- `axis = 0`: columnas


- `axis = 1`: filas

````python
# vamos a recordar lo que ten√≠amos en el arr1

arr1
````

````python
np.any(arr1)
````

````python
# como no estamos pasando ning√∫n eje nos devuelve un √∫nico valor. 
## si en alg√∫n caso se cumple la condici√≥n que le pasamos nos devolver√° True
### en este caso como tenemos valores en nuestro array mayores que 0.6 nos devolver√° True, aunque no sean todos

np.any(arr1 > 0.6)
````

````python
# ¬øse est√° cumpliendo la condici√≥n que pasamos en alguno de los valores
## de cada una de mis columnas? 

np.any(arr1 > 0.6, axis = 0)
````

Entendamos el resultado anterior: 

- En nuestra segunda columna tenemos los siguientes valores:
```
[0.72007867, 0.42533541, 0.35549331]
```
¬ø Es **alguno** de estos valores mayor que 0.6? SI. Entonces nos devuelve True


- ¬øQu√© pasa con la tercera columna? Tenemos los siguientes valores:

```
[0.06335547, 0.2394793, 0.41744516]
```

¬øEs **alguno** de esto valores mayores que 0.6? NO. Entonces nos devuelve False. 

````python
# lo podemos hacer igual con las filas
## en todas las filas de mi array tengo al menos un valor mayor que 0.6, 
### por eso nos devuelve todo True

np.any(arr1, axis = 1)
````

## `np.where()`:

```python

- El primero *array* del *output* hace referencia a las filas

```

- El segundo *array* hace refencia a las columnas

```python
array([0, 3, 2, 1])
```

Entonces ¬øqu√© significa?

- El elemento 0, 0 (fila, columna) cumple la condici√≥n


- El elemento 0, 3 (fila, columna) cumple la condici√≥n


- El elemento 1, 2 (fila, columna) cumple la condici√≥n


Y as√≠ sucesivamente, haciendo la combinaciones entre los dos *arrays* que nos devuelve.  

# Nota sobre los ejes (par√°metro `axis`)

En este jupyter hemos hecho todos los ejemplos con un *array* de dos dimensiones. Pero... ¬øc√≥mo podr√≠amos acceder a los distintos ejes en un *array* tridimensional? 


Veamos el siguiente diagrama: 

![ejes-tridimensional.jpeg](https://github.com/Adalab/data_imagenes/blob/main/Modulo-2/NumPy/index-tres-dimensiones.png?raw=true)


Si nos fijamos: 

- `axis = 0`: hace referencia a la dimensi√≥n del *array* 


- `axis = 1`: hace referencia a las columnas dentro de cada *array* 


- `axis = 2`: hace referencia a las filas dentro de cada *array* 

````python
# tenemos un array con dos matrices de tres filas y 5 columnas
arr3 = np.random.random_sample((2,3,5))
arr3
````

````python
# hagamos una suma
np.sum(arr3)
````

````python
# va a calcularnos la suma para cada matriz que tengamos en nuestro array

np.sum(arr3, axis = 0)
````

````python
## hagamos la suma para cada columna
## tendremos 10 resultados, uno por cada columna

np.sum(arr3, axis = 1)
````

````python
# hagamos la suma para cada fila
## tendremos 6 resultados. Uno por cada fila que tenemos en nuestro array

np.sum(arr3, axis = 2)
````
