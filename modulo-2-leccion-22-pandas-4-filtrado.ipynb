## Filtrado en base al valor de una columna

La operaci√≥n b√°sica de filtrado es seleccionar aquellos valores que cumplen una condici√≥n. Para ello se puede comparar la columna con el valor deseado y obtener una lista de valores verdaderos o falsos. 

Para este tipo de operaciones utilizaremos los operadores de comparaci√≥n, es decir, `==`, `!=`, `>`, `<`, `>=`, `<=`. 


Imaginemos que queremos extraer solo la informaci√≥n para los clientes que vienen de un determinado pais. 

````python
# antes de hacer el filtro queremos saber cuantos elementos √∫nicos tenemos en la columna "state"

df["state"].unique()
````

````python
# creemos el filtro para sacar solo los clientes de Nevada. Como siempre lo podremos almacenar en una variable para 
# poder trabajar con el resultado m√°s adelante

df_nevada = df[df["state"] == "Nevada"]
````

````python
# veamos que tenemos en df_nevada
df_nevada.head(2)
````

````python
# cuantos elementos √∫nicos tenemos ahora en la columna state?

print(f"En la columna 'state' solo tenemos este valor {df_nevada.state.unique()}")
print("-----------------------------------------------------------------------------------")
````

Lo que ha pasado es que hemos extraido unicamente aquellas filas donde el valor de la columna `state` es igual a Nevada. 

Esto tambi√©n lo podemos hacer con columnas que sean num√©ricas, por ejemplo la columna `total_claim_amount`. Imaginamos que queremos extraer los clientes que tienen un `total_claim` mayor que 500. 

````python
df_500 = df[df["total_claim_amount"] > 500]
````

````python
# veamos que tenemos en df_500. Ahora solo tendremos aquellas filas donde "total_claim_amount" tenga valores superiores a 500

print(f"En la columna 'total_claim_amount' solo tenemos este valor {df_500.total_claim_amount.unique()}")
print("-----------------------------------------------------------------------------------")
````

##  Filtrado en base al valor de m√∫ltiples valores

Hasta ahora hemos visto como filtrar por una √∫nica condici√≥n, pero podemos incluir varias condiciones, bas√°ndonos en una √∫nica columna o varias. 

Los operadores l√≥gicos que vamos a usar para hacerlo son: 

- `&`: es el `and` que conoc√≠amos hasta ahora. Se tienen que cumplir las dos condiciones que pasamos
- `|`: es el `or` que conoc√≠amos hasta ahora. Se tiene que cumplir una condici√≥n u otra. 
- `~`: operador `not`. Se usa para negar una condici√≥n.

````python
# veamos un ejemplo del operador `~` para entenderlo mejor

valores = pd.Series([True, True, False])
valores
````

````python
# si nos fijamos, el operador ~ nos esta devolviendo el opuesto de los datos que tenemos en valores. 
~valores
````

````python
df.head(2)
````

**Filtrando usando el operador `&`**

Podemos incluir tantas condiciones como queramos, por ejemplo, buscar aquellos clientes que sean de Arizona, que tengan trabajo y sean mujeres. 

‚ö†Ô∏è Notad que cada condici√≥n va en par√©ntesis!

````python
df_arizona_mujeres = df[(df["state"] == "Arizona") & (df["employmentstatus"] == "Employed") & (df["gender"] == "F")]
df_arizona_mujeres.head(2)
````

````python
print(f"En la columna 'state' solo tenemos este valor {df_arizona_mujeres.state.unique()}")
print("-----------------------------------------------------------------------------------")
print(f"En la columna 'employmentstatus' solo tenemos este valor {df_arizona_mujeres.employmentstatus.unique()}")
print("-----------------------------------------------------------------------------------")
print(f"En la columna 'gender' solo tenemos este valor {df_arizona_mujeres.gender.unique()}")
````

**Filtrando usando el operador `|`**

Vamos a buscar a aquellos clientes que sean de Nevada o de Arizona

````python
df_arizona_nevada = df[(df["state"] == "Arizona") |(df["state"] == "Nevada") ]
df_arizona_nevada.head(2)
````

````python
print(f"En la columna 'state' solo tenemos este valor {df_arizona_nevada.state.unique()}")
print("-----------------------------------------------------------------------------------")
````

**Filtrando usando el operador `&` y `|`**

Incluso podemos buscar usando los dos operadores. Buscaremos los clientes que sean de Arizona o Nevada y que sean solteros. 

````python
filtro_combinado = df[((df["state"] == "Arizona") |(df["state"] == "Nevada")) & ((df["marital_status"] == "Single"))]
filtro_combinado.head(2)
````

````python
print(f"En la columna 'state' solo tenemos este valor {filtro_combinado.state.unique()}")
print("-----------------------------------------------------------------------------------")
print(f"En la columna 'marital_status' solo tenemos este valor {filtro_combinado.marital_status.unique()}")
print("-----------------------------------------------------------------------------------")
````

**Filtrando usando el operador `~`**

Este operador lo usaremos para negar una condici√≥n. Por ejemplo, la condici√≥n que hemos establecido de buscar los clientes de Nevada y Arizona: 

````python
# este es el filtro que aplicamos 
(df["state"] == "Arizona") |(df["state"] == "Nevada")
````

En este caso vamos a querer filtrar aquellos clientes que NO sean de Nevada o Arizona. 

````python
# si le ponemos el operador ~ antes que pasa? NOS DA LA VUELTA A LA CONDICION
~(df["state"] == "Arizona") |(df["state"] == "Nevada")
````

````python
# apliquemos el filtro para nos devuelva los datos
filtro = (df["state"] == "Arizona") |(df["state"] == "Nevada")
filtro_inverso = df[~filtro]
filtro_inverso.head(2)
````

````python
print(f"En la columna 'state' solo tenemos este valor {filtro_inverso.state.unique()}")
print("-----------------------------------------------------------------------------------")
````

Nos aparecen los clientes que **NO** sean de Arizona o Nevada. 

## Filtrado usando m√©todos especificos de Pandas 

En Pandas tambi√©n tenemos una serie de m√©todos que nos va a permitir filtrar nuestros datos, algunos de los m√°s importantes son: 

- `isin()`
- `str.contains()`
- `notnull()`

### `isin()`

Si cada elemento del *DataFrame* o una columna espec√≠fica contiene valores que especificamos. 

La sintaxis: 

```python
DataFrame.isin(values)
```

Donde: 

- `values`: puede ser un iterable, una Serie, un DataFrame o un diccionario. 

````python
df.head(2)
````

**Ejemplo con una lista**

Filtramos por aquellos clientes cuyo nivel educativo sea Doctor o Master. Lo que haremos ser√° crearnos una lista con los valores que nos interesan para luego pas√°rselo al m√©todo `isin()`

````python
df_educacion = df[df["education"].isin(["Doctor", "Master"] )]
````

````python
print(f"En la columna 'education' solo tenemos este valor {df_educacion.education.unique()}")
print("-----------------------------------------------------------------------------------")
````

**Ejemplo con una Serie**

Vamos a crear una serie a partir de un diccionario, donde las *keys* ser√°n el √≠ndice de nuestra serie y har√° referencia a la columna sobre la que queremos aplicar el filtro y los *values* ser√°n los valores que queremos filtrar.  

````python
# creamos una variable que ser√° una serie con el √≠ndice como la columna de interes y los valores lo que queremos filtrar

values = pd.Series({"education": "Master", "education": "Doctor"})
````

````python
df_educacion_isin = df[df["education"].isin(values)]
````

````python
df_educacion_isin.head(2)
````

````python
print(f"En la columna 'education' solo tenemos este valor {df_educacion_isin.education.unique()}")
print("-----------------------------------------------------------------------------------")
````

### `str.contains()` 

Comprueba si un patr√≥n o regex est√° contenido en una cadena de una columna especificada. 

La sintaxis que debemos seguir: 

```python
MiColumna.str.contains(pat, case=True, regex=True)
```

Donde: 

- `pat`: nuestro caracter o expresi√≥n regular. 
- `case`: por defecto True. 

    - True: es *case sensitive*
    - False: no es *case sensitive* 
    
- `regex`: por defecto True. 

    - True: asume que el `pat` es una expresi√≥n regular
    - False: trata el `pat` como una cadena literal 

````python
# antes de empezar a aplicar este m√©todo recordemos el DataFrame

df.head(2)
````

````python
# si exploramos los valores √∫nicos de la columna "vehicle_class" si quisieramos extraer los veh√≠culos de categor√≠a
# Luxury tendr√≠amos que hacer dos filtros o un filtro m√∫ltiple. 

# Sin embargo con "str.contains" podemos hacerlo de una forma m√°s sencilla, usando patrones de regex. 

df.vehicle_class.unique()
````

````python
# lo primero que hacemos es definir nuestro patr√≥n de regex. Para recordar, en este patr√≥n estamos diciendo todas
# aquellas palabras que tengan un "L" seguido de m√°s letras

patron = 'L\w+' 


# tambi√©n podemos pasar la palabra literal
patron2 = "Luxury"
````

````python
clase_luxury = df[df["vehicle_class"].str.contains(patron, regex=True)]
clase_luxury.head(2)
````

````python
# si ahora chequeamos los valores √∫nicos del DataFrame clase_luxury solo tendremos dos valores. Solo los que incluyen
# "luxury"

print(f"En la columna 'vehicle_class' solo tenemos este valor {clase_luxury.vehicle_class.unique()}")
print("-----------------------------------------------------------------------------------")
````

````python
# Si aplicamos el segundo patr√≥n que ten√≠amos

clase_luxury2 = df[df["vehicle_class"].str.contains(patron2, regex=True)]
clase_luxury2.head(2)
````

````python
# chequeamos los resultados. Deber√≠an ser iguales a los obtenidos cuando usamos el patron

print(f"En la columna 'vehicle_class' solo tenemos este valor {clase_luxury2.vehicle_class.unique()}")
print("-----------------------------------------------------------------------------------")
````

# Reemplazar valores bas√°ndose en √≠ndices y condiciones

En este apartado vamos a tener que recordar conceptos que aprendimos en las lecciones anteriores: 

- `index`: nos va a devolver el √≠ndice de los elementos de nuestro DataFrame.
- `iloc`: lo usamos para acceder a elementos de nuestro DataFrame bas√°ndonos en su posici√≥n. 

En este apartado vamos a trabajar con la columna `vehicle_class`. ¬øQu√© es lo que haremos? En aquellos clientes que tengamos un "SUV" reemplazaremos ese valor por la definici√≥n completa, es decir, `Sport Utility Vehicle`.

````python
# lo primero que vamos a hacer es sacar los √≠ndices bas√°ndonos en una condici√≥n. En este caso que el tipo de veh√≠culo 
# sea igual a SUV. 

indices = df.index[(df["vehicle_class"] == "SUV")]
````

````python
# estos son los √≠ndices de las filas donde se cumple la condici√≥n

indices
````

````python
# ahora vamos a necesitar acceder a cada elemento de nuestra variable indices. 
# Para eso usaremos un for

for ind in indices: 
    
    # una vez que tenemos cada elemento lo que tenemos que hacer es usar el iloc para acceder 
    # a los elementos y reemplazarlos
    df["vehicle_class"].iloc[ind] = "Sport Utility Vehicle"
````

````python
df.head(4)
````

````python
# chequemos los resultados. Ya no tenemos SUV, si no 'Sport Utility Vehicle'

print(f"En la columna 'vehicle_class' solo tenemos este valor {df.vehicle_class.unique()}")
print("-----------------------------------------------------------------------------------")
````

# Reemplazar valores bas√°ndose en m√©todos de NumPy

Desde Pandas, podemos usar m√©todos de NumPy para reemplazar valores de nuestro *DataFrame*, estos m√©todos son: 

- `np.where()`
- `np.select()`

## `np.where()`

Este m√©todo nos va a permitir reemplazar los elementos que cumplan una condici√≥n dada por los valores que especifiquemos. En caso de que no se cumpla la condici√≥n podremos indicar por qu√© queremos reemplazar ese elemento. 

La sintaxis que seguiremos en este m√©todo es:

```python
np.where(cond, true, false)
```

Donde:

- `cond`: es una condici√≥n que nosotros querramos que cumpla el DataFrame.
- `true`: es el valor que queremos que se reemplace en el DataFrame cuando *s√≠* se cumpla la condici√≥n.
- `false`: es el valor que queremos que se reemplace en el DataFrame cuando *no* se cumpla la condici√≥n.


````python
# recordemos el DataFrame

df.head(2)
````

En este caso vamos a trabajar con la columna `number_of_policies`. Vamos a establecer un umbral por encima del cual consideraremos que es un cliente fiel, y en caso contrario no ser√° fiel.

Lo que haremos ser√° clasificar a nuestros clientes en "supreme" si tienen m√°s de 6 polizas o "good" si tienen menos.

````python
# en este c√≥digo lo que estamos especificando es que si el n√∫mero de polizas es mayor que 6, 
# en la nueva columna "new_number_policies" le asignes una valor de "supreme" y en caso de 
# que no se cumpla la condici√≥n nuestro cliente ser√° asignado como "good"

df["new_number_policies"] = np.where(df["number_of_policies"] > 6,
                            "supreme", "good")
````

````python
df.head(2)
````

````python
print(f"En la columna 'new_number_policies' solo tenemos este valor {df.new_number_policies.unique()}")
print("-----------------------------------------------------------------------------------")
````

## `np.select()`

Devuelve un array extra√≠do de los elementos de choicelist, dependiendo de las condiciones.

La sintaxis en este caso: 

```python
np.select(condlist, choicelist)
```

Donde:

- `condlist`: ser√° una lista con las condiciones que queremos chequear. 
- `choicelist`: ser√° otra lista con lo que queremos que se devuelva en funci√≥n de las condiciones definidas anteriormente. 

üìå Es bastante similar a `np.where`, pero con la ventaja de que en `np.select` podemos incluir m√°s de dos condiciones. 

Apliquemos las mismas condiciones que en el `np.where`. Clientes con m√°s de 6 polizas ser√°n catalogados como cliente "supreme" y clientes con menos polizas ser√°n "good". 

````python
# creamos la lista de las dos condiciones, que sea mayor o menor que 6
condiciones = [df["number_of_policies"] > 6, df["number_of_policies"] < 6]

# creamos la lista de los valores que querremos crear cuando se cumpla cada condici√≥n
opciones = ["supreme", "good"]

# aplicamos el m√©todo a nuestro DataFrame y creamos una nueva columna llamada "new_select"
df["new_select"] = np.select(condiciones, opciones)
````

````python
df.head(2)
````

Usemos este m√©todo en otra columna e incluimos m√°s de dos condiciones. En este caso vamos a trabajar con la columna `policy_type` que tiene los siguientes valores √∫nicos:

```python
df.policy_type.unique()
array(['Corporate Auto', 'Personal Auto', 'Special Auto'], dtype=object)
```

En este caso vamos a querer quitar por ejemplo la palabra "Auto" de todos los elementos. Veamos como hacerlo.

````python
# primero creamos las condiciones

condiciones2 = [df['policy_type'] == 'Corporate Auto',
                df['policy_type'] == 'Personal Auto' ,
                df['policy_type'] == 'Special Auto' ]

# definimos los valores que querremos poner

opciones2 = ['Corporate', 'Personal', 'Special']
````

````python
# por √∫ltimo aplicamos el m√©todo np.select creando una columna nueva que se llame "policy_type_new". 

df["policy_type_new"] = np.select(condiciones2, opciones2)
````

````python
# chequeamos nuestro DataFrame
df.head(2)
````

````python
print(f"En la columna 'policy_type_new' solo tenemos este valor {df.policy_type_new.unique()}")
print("-----------------------------------------------------------------------------------")
````
