# `concat()`

Esta es la funci칩n que deber칤amos usar si tenemos varios *dataframes* con los mismos nombres de columnas. 

Podemos utilizar la funci칩n `concat` para concatenar columnas o filas de un *dataframe* a otro.

Sintaxis: 

```python
pd.concat(objs, axis=0, join='outer', ignore_index=False)
```
- `objs`: una **lista** con los nombres de los *dataframes* que queremos concatenar


- `axis`: el eje sobre el que queremos aplicar la operaci칩n. Por defecto 0

    - `axis=0` le dice a pandas que apile el segundo DataFrame DEBAJO del primero. Es decir, estamos concatenando por **columnas**. Nos unir치 por lo nombres de las columnas que sean iguales. 

    >Cuando usamos este eje, tenemos que asegurarnos de que tenemos las mismas columnas y el mismo formato de columna asociado en ambos conjuntos de datos.En caso de que tengamos alguna columna que no este en los dos dataframes, nos pondr치 NaN en los valores para el *dataframe* que no ten칤a esa columna

    - `axis=1` apilar치 las columnas del segundo DataFrame a la DERECHA del primer DataFrame. Es decir, estamos concatenando por **filas** 

    > Cuando usamos este eje, queremos asegurarnos de que lo que estamos haciendo tiene sentido (es decir, los datos est치n relacionados de alguna manera).


- `join`: como queremos que sea la uni칩n. Por defecto *outer*, es decir, nos va a pasar la informaci칩n de las dos tablas. Puede ser tambi칠n *inner* que se quedar치 solo con aquellos elementos que aparezcan en las dos tablas. 


- `ignore_index`: por defecto es False. Si es True, no se utilizan los valores de 칤ndice a lo largo del eje de concatenaci칩n. El eje resultante ser치 etiquetado como 0, ..., n - 1. Esto es 칰til si est치 concatenando objetos donde el eje de concatenaci칩n no tiene informaci칩n de indexaci칩n significativa.

# `merge()` 

Probablemente la forma m치s com칰n de a침adir nuevas columnas a un *dataframes* basado en algunas condiciones comunes.


Sintaxis: 

```python
dataframe1.merge(dataframe2, how='inner', on, left_on=None, right_on=None)
```

Donde: 

- `dataframe2`: objeto que vamos a fusionar. 


- `how`: tipo de mergeo que vamos a hacer. Por defecto `inner`. Ve치moslo con m치s detalle: 

    - `left`: usa s칩lo las claves del *dataframe* izquierdo, similar a una uni칩n externa izquierda de SQL.

    - `right`: usa s칩lo las claves del *dataframe* derecho, similar a una uni칩n externa derecha de SQL.

    - `outer`: utiliza la uni칩n de las claves de ambos *dataframe*, es decir, me va a devolver todo. 
    
    - `inner`: utiliza la intersecci칩n de las claves de ambos *dataframe*, similar a una uni칩n interna de SQL.

    - `cross`: crea el producto cartesiano de ambos marcos, conserva el orden de las claves de la izquierda.

- `on`: columna o lista de columnas por las que queremos hacer la uni칩n. 


- `left_on` y `right_on`: son similares al par치metro `on`


**쯇or qu칠 hacemos esta diferenciaci칩n entre `on` , `left_on` y `right_on`?**

- Podemos usar `on` si los nombres de las columnas por los que queremos unir nuestros *datasets* se llaman igual 

- Por el contrario, si las columnas por las que vamos a mergear de nuestros *dataframes* tienen distintos nombres usaremos `left_on` y `right_on`.
Para entenderlo mejor por defecto, el `merge` es como un *inner join* de SQL, visualmente lo ver칤amos as칤: 



Para ilustrar el uso de merge() volveremos a nuestro ejemplo de las especies que ten칤a un marco de datos de muestreos de 2000 , 2001 y 2002. Pero en este caso vamos a incluir un nuevo fichero de datos, el de la informaci칩n de las especies. 


````python
df_species = pd.read_csv("species.csv", index_col = 0).reset_index()
df_species.head()
````

````python
# recordemos uno de los dataframes de muestreos 
df_2002.head(2)
````

````python
print(f"las filas y columnas que tenemos en el df_2002 son {df_2002.shape}")
print("-------------------------------------------------------------------")


print(f"las filas y columnas que tenemos en el df_species son {df_species.shape}")
print("-------------------------------------------------------------------")
````

Por otro lado, veamos los nombres de columnas que tenemos en los *dataframes* 

````python
print(f"los nombres de columnas que tenemos en el df_2002 son \n\n {df_2002.columns}")
print("-------------------------------------------------------------------")


print(f"los nombres de columnas que tenemos en el df_species son \n\n {df_species.columns}")
print("-------------------------------------------------------------------")
````

Si nos fijamos, tenemos 5 columnas diferentes a las que ten칤amos en los dataframes, peeeero hay una columna que comparten! La de `species_id`. 

Esta columna ser치 la que nos permita hacer el mergeo entre los dos *dataframes* 

## Inner merge 

Como tenemos una columna en com칰n que se llama igual en los *dataframes* usaremos el par치metro `on`

````python
mergeado_inner = df_2002.merge(df_species, on='species_id')
mergeado_inner.head()
````

````python
print(f"las filas y columnas que tenemos en el df_2002 son {df_2002.shape}")
print("-------------------------------------------------------------------")


print(f"las filas y columnas que tenemos en el mergeado_inner son {mergeado_inner.shape}")
print("-------------------------------------------------------------------")

````

Fijaos que ahora tenemos nuestros *datasets* unidos, se nos han juntado las columnas de los *dataframes* en aquellos valores que coinciden los valores de `species_id`, que es por donde hemos especificado que mergee. 

> Tenemos menos filas en el *dataframe* mergeado. Esto ser치 porque hay c칩digo que est치n en df_2002 y no en df_species. Como no encuentra el "match" los elimina. 
>
>El *dataframe* mergeado_inner resultante contiene todas las columnas de df_2002 (record_id, month, day	year, etc.) as칤 como todas las columnas de df_species (species_id, genus, species, taxa).
>
>Tambi칠n observamos que mergeado_inner tiene menos filas que df_2002. Esto indica que hab칤a filas en df_2002 con valores para species_id que no existen como valores para species_id en df_species.

````python
# 游늷 esto es lo mismo que lo anterior: 

mergeado_inner2 = df_2002.merge(df_species, left_on = "species_id", 
                               right_on = "species_id")
mergeado_inner2.head()
````

## Left merge

Pero usando `merge` podemos hacer otro tipo de uniones en nuestros *dataframes* como el *left join* 


````python
mergeado_izq = pd.merge(left=df_2002, right=df_species, how='left',
                        left_on='species_id', right_on='species_id')

mergeado_izq.head()
````

````python
print(f"las filas y columnas que tenemos en el df_2002 son {df_2002.shape}")
print("-------------------------------------------------------------------")


print(f"las filas y columnas que tenemos en el mergeado_inner son {mergeado_izq.shape}")
print("-------------------------------------------------------------------")

````

El DataFrame resultante de una *left join* se parece mucho parece mucho al resultado del *inner join*  en cuanto a las columnas que contiene. Sin embargo, a diferencia del *inner_join* (mergeado_inner), *left_join* (mergeado_izq) contiene el mismo n칰mero de filas que el *dataframe* original de df_2002. 

Cuando examinamos mergeado_izq, encontramos que hay filas en las que falta la informaci칩n que deber칤a proceder de df_species (es decir, species_id, genus y taxa) (contienen valores NaN):

> Tenemos 37 NaN en las columnas species_id, genus, species, taxa del *dataframe* df_species. Esto significa que hay 37 especies que ten칤amos en el df_2002 que NO aparec칤an en df_species. 

````python
mergeado_izq.isnull().sum()
````

# `join()`

Este es un caso especial de fusi칩n cuando al menos una de las columnas sobre las que se est치 uniendo es un 칤ndice. 

Basicamente es lo mismo, pero si queremos usar `join`, la uni칩n la tenemos que hacer por los 칣ndices. No hace falta que las columnas a fusionar est칠n sean 칣ndices, pero **al menos una** debe serlo.

Sintaxis: 

```python
dataframe1.join(dataframe2, on , how = "left" , lsuffix='', rsuffix='')
```

Donde: 

- `on`: la columna o el 칤ndice por el que queremos hacer la uni칩n. 


- `how`: por defecto `left`, pero podemos especificar *right*, *outer* o *inner*. 


- `lsuffix`: por defecto nada, en caso de que coincidan columnas en los dataframes que queremos unir, ser치 el sufijo que aparecer치 en esas columnas en el dataframe izquierdo



- `rsuffix`: por defecto nada, en caso de que coincidan columnas en los dataframes que queremos unir, ser치 el sufijo que aparecer치 en esas columnas en el dataframe derecho


Volvamos sobre nuestro ejemplo recordando sus formas:

````python
print(f"las filas y columnas que tenemos en el df_2002 son {df_2002.shape}")
print("-------------------------------------------------------------------")


print(f"las filas y columnas que tenemos en el mergeado_inner son {df_species.shape}")
print("-------------------------------------------------------------------")

````

Lo primero que tenemos que hacer es establecer en alguno de nuestros *dataframes* la columna de `species_id` (que es la que comparten) como 칤ndice. 

````python
# recordamos que para poner una columna como 칤ndice tenemos que usar el m칠todo "set_index"

df_species.set_index(["species_id"], inplace = True)
````

````python
# veamos como ha quedado el dataframe. Bravo 游녪游낗!ya tenemos nuestro 칤ndice.  

df_species.head(2)
````

````python
# como en este caso no tenemos columnas coincidentes a parte de las que vamos a usar para 
## juntar nuestros dataframes no le pasaremos los par치metros lsuffix y rsuffix. 

df_join = df_2002.join(df_species, on = "species_id")
df_join.head()
````

````python
print(f"las filas y columnas que tenemos en el df_2002 son {df_2002.shape}")
print("-------------------------------------------------------------------")


print(f"las filas y columnas que tenemos en el mergeado_inner son {df_join.shape}")
print("-------------------------------------------------------------------")

````
